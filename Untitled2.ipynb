```python
import requests
import pandas as pd
import time

BASE_URL = "http://openapi.seoul.go.kr:8088/6c4e55755068616e35306566524b4e/json/energyUseDataSummaryInfo"

def get_year_months():
    year_months = []
    for year in range(2015, 2016):
        for month in range(1, 13):
            year_months.append(f"{year}{month:02d}")
    return year_months

responses = []

for ym in get_year_months():
    year = ym[:4]
    month = ym[4:]
    url = f"{BASE_URL}/1/1000/{year}/{month}"

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        info = data.get("energyUseDataSummaryInfo")
        rows = info.get("row")
        responses.extend(rows)
        print("api 호출 성공")
        print(response.json())

    else:
        print(f"API 호출 실패: {response.status_code}")

df = pd.DataFrame(responses)

# 전체 데이터 중에서 필요한 컬럼만 선택
df = df[["MM_TYPE", "YEAR", "MON", "EUS", "GUS","WUS","HUS" ]]

# 개인 유형 + 현년만
df = df[(df["MM_TYPE"] == "개인") ]

season_dict = {
    "12": "겨울", "01": "겨울", "02": "겨울",
    "03": "봄", "04": "봄", "05": "봄",
    "06": "여름", "07": "여름", "08": "여름",
    "09": "가을", "10": "가을", "11": "가을"
}

df["season"] = df["MON"].map(season_dict)

df

```
